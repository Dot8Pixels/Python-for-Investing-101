```python
import pandas as pd
import pandas_ta as ta
import numpy as np
import yfinance as yf
import vectorbt as vbt
import datetime as datetime
```

```python
data = pd.DataFrame()
data = data.ta.ticker('AAPL', period='3y', interval='1d')
```

```python
df = data.copy()
df.head()
```

```python
custom_strategy = ta.Strategy(
    name = 'Custom STA',
    description = 'For Testing',
    ta = [
        {'kind':'sma', 'length':20},
        {'kind':'ema', 'length':5}
    ]
)
```

```python
df.ta.strategy(custom_strategy)
```

```python
df.tail()
```

```python
def my_strategy(df):
    if (df.Close < df.EMA_5) and (df.Close < df.SMA_20) :
        return True
    else :
        return False
```

```python
df['signal'] = df.apply(my_strategy,axis=1).vbt.fshift(1)
df
```

```python
df = df.iloc[1:,:] # Remove NaN
df
```

```python
signal_vectorbt = df.ta.tsignals(df.signal, asbool=True, append=True)
signal_vectorbt
```

```python
port = vbt.Portfolio.from_signals(df.Close,
                                  direction='both',
                                  entries = signal_vectorbt.TS_Entries,
                                  exits = signal_vectorbt.TS_Exits,
                                  freq = 'D',
                                  init_cash = 100000,
                                  fees = 0.0025,
                                  slippage = 0.0025)
```

```python
port.plot().show()
```

```python
custom_strategy2 = ta.Strategy(
    name = 'Custom STA2',
    description = 'For Testing',
    ta = [
        {'kind':'rsi', 'length':14},
        {'kind':'macd'}
    ]
)
```

```python
df2 = data.copy()
df2.ta.strategy(custom_strategy2)
df2
```

```python
def my_strategy2(df):
    if (df.RSI_14 < 80) and (df.MACD_12_26_9 > df.MACDs_12_26_9):
        return True
    else:
        return False
```

```python
df2['signal'] = df2.apply(my_strategy2, axis=1)
df2
```

```python
signal_vectorbt2 = df2.ta.tsignals(df2.signal, asbool=True, append=True, trade_offset=1) # trade_offset; backtest use 1, live use 0
```

```python
port2 = vbt.Portfolio.from_signals(df2.Open,
                                  direction='both',
                                  entries = signal_vectorbt2.TS_Entries,
                                  exits = signal_vectorbt2.TS_Exits,
                                  freq = 'D',
                                  init_cash = 100000,
                                  fees = 0.0025,
                                  slippage = 0.0025)
```

```python
port2.plot().show()
```
