```python
data = pdr.get_data_yahoo('AAPL', start='2017-01-01', end='2021-08-05')
data.head()
```

```python
sma_50 = df.rolling(50).mean()
sma_100 = df.rolling(100).mean()
sma_200 = df.rolling(200).mean()
sma_50
```

```python
sma_200.plot()
```
