```python
df5['52-week-high'] = df5.groupby('Stock')['Close'].transform('max')
df5['52-week-low'] = df5.groupby('Stock')['Close'].transform('min')
```

```python
df5['from-high'] = df5['Close']/df5['52-week-high']-1
```

```python
df5
```

```python
df5.loc[(df5['RSI_14'] > 50) & 
        (df5['MACDh_12_26_9'] > 0) & 
        (df5['MACD_12_26_9'] > 0) &
        (df5['from-high'] > -0.2), 'Action'] = 'Hodl'
df5.loc[(df5['RSI_14'] > 50) & 
        (df5['MACDh_12_26_9'] < 0) & 
        (df5['from-high'] > -0.2), 'Action'] = 'Aware'
```
